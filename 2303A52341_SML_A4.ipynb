From the above data:

Read the data with pandas and find features and target variables
Normalize the data with min-max scaling

```python
import pandas as pd
data=pd.read_csv("/content/train.csv")
print(data)
```

```python
y=data['price_range']
y
```

```python
x=data.drop('price_range',axis=1)
x
```

```python
new=data-x.min()/x.max()-x.min()
new
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(new,y,test_size=0.2,random_state=30)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```
